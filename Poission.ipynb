```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Parameters
 = 5  # average rate (calls per hour)
k_values = np.arange(0, 16)  # range of calls from 0 to 15

# Poisson PMF function
def poisson_pmf(k, ):
    return (math.exp(-) * **k) / math.factorial(k)

# Calculate probabilities for each k
probabilities = [poisson_pmf(k, ) for k in k_values]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')

# Highlight k = 3
bars[3].set_color('orange')

# Labels and title
plt.title(f'Poisson Distribution ( = {}) - Probability of k Calls per Hour')
plt.xlabel('Number of Calls (k)')
plt.ylabel('Probability P(X = k)')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
