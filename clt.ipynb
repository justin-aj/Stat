```python
# CENTRAL LIMIT THEOREM
```

```python
import numpy as np

a = np.random.randint(0, 100, size=300)
```

```python
c = []
for i in range(1000):
    b = np.random.choice(a, size=20, replace=False)
    c.append(b.mean()) 
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(c, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribution of Sample Means (1000 Samples)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
