```python
import numpy as np
data = np.random.randint(0, 10, 1000)
```

```python
c = []
for i in range(1000):
    b = np.random.choice(data, size=20, replace=False)
    c.append(b.mean()) 
```

```python
data = c
```

```python
mean = sum(data) / len(data)
print(f"Mean: {mean}")
```

```python
# Manual calculation
squared_diffs = [(x - mean) ** 2 for x in data]
variance_population = sum(squared_diffs) / len(data)  # Divide by n
variance_sample = sum(squared_diffs) / (len(data) - 1) # Divide by n-1

print(f"Population Variance: {variance_population}")
print(f"Sample Variance: {variance_sample}")
```

```python
import math

sd_population = math.sqrt(variance_population)
sd_sample = math.sqrt(variance_sample)

print(f"Population SD: {sd_population}")
print(f"Sample SD: {sd_sample}")
```
